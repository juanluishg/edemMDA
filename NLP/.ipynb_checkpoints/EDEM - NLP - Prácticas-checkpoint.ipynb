Ahora tendrás la oportunidad de escribir regex para que coincidan con dígitos, strings y caracteres no alfanuméricos. 

Nota: Es importante añadir a tus patrones regex, el prefijo con la letra `r` para asegurarte de que los patrones se interpretan de la forma que quieres. De lo contrario, puedes encontrar problemas relacionados con las [secuencias de escape](https://tutz.tv/python/secuencias-de-escape/) en las cadenas. Por ejemplo, "\n" en Python se utiliza para indicar una nueva línea, pero si utilizas el prefijo r, se interpretará como el *raw string* "\n" - es decir, el carácter "\\" seguido del carácter "n" - y no como un salto de línea.

Recuerda de las transparencias de la sintaxis de la librería regex, que **primero se le pasa el pattern y segundo el string**.

```python
#Esta es la frase que vamos a utilizar para esta primera parte de las prácticas.
my_string = " Don Quijote soy, y mi profesión la de andante caballería. Son mis leyes, el deshacer 1000 entuertos, prodigar el bien y evitar el mal. Huyo de la vida regalada, de la ambición y la hipocresía, y busco para mi propia gloria la senda 33 más angosta y difícil. ¿Es eso, de tonto y mentecato?"
my_string
```

Vamos a devolver distintos patrones dentro de la frase quijotesca que hemos visto arriba. En este caso mostraremos por pantalla 3 resultados diferentes.




1.   Dividiremos la frase cuando encontremos un punto, coma o signo de exclamación/interrogación
2.   Encontrar todas las mayúsculas en la Frase.
3.   Dividir la frase en espacios e imprimir el resultado



```python
# Escribimos un pattern para encontrar finales de frase (.?!).
sentence_endings = r"[.?!]"

# Dividir my_string en los finales de frase e imprimir el resultado
print(re.split(sentence_endings, my_string))

# Encontrar todas las mayúsculas en my_string. Recuerda el patrón [a-z] que hemos visto en las slides. 
capitalized_words = r"[A-ZÑ]\w+"
print(re.findall(capitalized_words, my_string))

# Divide my_string en espacios e imprime el resultado
spaces = r"\s+"
print(re.split(spaces, my_string))

# Encuentra todos los dígitos en my_string
digits = r"\d+"
print(re.findall(digits, my_string))
```

### 1.2 Ejercicio 1: Expresiones Regulares

Encuentra todos los mails posibles en el siguiente string y muéstralos por pantalla


```python
message = """Contacta con nosotros en el siguiente correo:  info@host.com.
    Si tienes cualquier problema técnico, escríbenos a sat@host.com
    si quieres contactar conmigo escríbeme a marin-melchorm@gmail.com
    """

pattern = r"[\w+\.-]+\@[\w\.-]+" 

mails = re.findall(pattern, message)


for mail in mails:
    print(mail)
```

### 2.2 Regex con tokenización NLTK

Twitter es un recurso bastante utilizado para textos y tareas de NLP. En este ejercicio, construirás un tokenizador más complejo para tweets con hashtags y menciones usando `nltk` y regex. La clase `nltk.tokenize.TweetTokenizer` te ofrece algunos métodos y atributos adicionales para analizar tweets.

En este caso, se te dan algunos tweets de ejemplo para analizar utilizando tanto `TweetTokenizer` como `regexp_tokenize` del módulo `nltk.tokenize`. 

A diferencia de la sintaxis de la biblioteca regex, con `nltk_tokenize()` se pasa el patrón como segundo argumento.


```python
# Importamos los módulos
from nltk.tokenize import regexp_tokenize
from nltk.tokenize import TweetTokenizer
import pandas as pd

#Cargamos la lista de tweets
lista_tuits = pd.read_csv('list_tweets.csv')
print(lista_tuits)
```

```python
#Convertimos en una lista
lista_tuits = lista_tuits['tweets'].to_list()

# Definimos un patrón regex
pattern1 = r"#\w+"

#Elegimos el primer tweet de la lista y lo mostramos
tweet = str(lista_tuits[0])
print(tweet)
```

```python
#Utilizamos el patrón regez y se lo pasamos al regexp_tokenize para extraer los hashtags
hashtags = regexp_tokenize(tweet, pattern1)
print(hashtags)

```

```python
# Escribe un patrón que coincida tanto con las menciones (@) como con los hashtags (#) 

#Para ello, utilizaremos el segundo tweet de la lista

tweet2 = lista_tuits[1]
print(tweet2)



```

```python
pattern2 = r'[\#|\@]\w+'


# Utiliza el patrón del último tuit, en la lista de tuits.
mentions_hashtags = regexp_tokenize(tweet2, pattern2)
print(mentions_hashtags)
```

```python
# Utilice el TweetTokenizer para tokenizar todos los tweets en una lista

tknzr = TweetTokenizer()


#Para hacer esto, utiliza el método .tokenize() de tknzr, donde t es el iterador
all_tokens = [tknzr.tokenize(t) for t in lista_tuits]
print(all_tokens)
```

### 2.4 Gráficos de longitud de palabra con matplotlib()

Matplotlib es una librería de gráficos ampliamente utilizada en Python para crear visualizaciones de datos, gráficos y diagramas. Tiene una funcionalidad bastante sencilla con muchas opciones para gráficos como histogramas, gráficos de barras, gráficos de líneas y scatter plots. Incluso tiene funcionalidades avanzadas como la generación de gráficos 3D y animaciones.



>El uso de `comprehension lists` acelerará los cálculos. Por ejemplo: `my_lines = [tokenize(l) for l in lines]` llamará a una función `tokenize` en cada línea de la lista `lines`. La nueva lista transformada se guardará en la variable `my_lines`.

```python
#Primero, vamos a cargar el texto. Ahora mismo volvemos a utilizar la primera escena de la película los caballeros de la mesa cuadrada.
nltk.download('webtext')
from nltk.corpus import webtext
import matplotlib.pyplot as plt
monty_python = webtext.raw('grail.txt')

#Imprimimos los 100 primeros carácteres
print(monty_python[0:100])

```

```python
# Divide el guión de monty_python en saltos de lineas usando el carácter \n
lines = monty_python.split('\n')

# Utiliza r.sub() dentro de una lista de comprehension  para reemplzar las indicaciones de guión
#como ARTHUR: y SOLDIER #1. El pattern esta vez está escrito :)
pattern = "[A-Z]{2,}(\s)?(#\d)?([A-Z]{2,})?:"
lines = [re.sub(pattern, '', l) for l in lines]

# Tokeniza cada linea en una lista de comprehensión manteniendo SOLO PALABRAS. Recuerda el patrón.
tokenized_lines = [regexp_tokenize(s, r"\w+") for s in lines]

# Utiliza una comprehension list para crear una lista de lineas llamada line_num_words.
#En este caso, el iterador es t_line
line_num_words = [len(t_line) for t_line in tokenized_lines]

# Realiza un histograma con matplotlib
plt.hist(line_num_words)

plt.show()
```

Como se aprecia en el histograma, la mayor parte de las palabras tienen una longitud de entre 1 y 10 carácteres.

```python
with open('/content/drive/MyDrive/Edem/NLP/english_stopwords.txt', 'r') as file:
    english_stops = file.read()
```

```python
import glob

path_list = glob.glob('/content/drive/MyDrive/Edem/NLP/Wikipedia articles/*.txt')
articles = []
for article_path in path_list:
    article = []
    with open(article_path, 'r') as file:
        a = file.read()
    tokens = word_tokenize(a)
    lower_tokens = [t.lower() for t in tokens]
    
    # Retain alphabetic words: alpha_only
    alpha_only = [t for t in lower_tokens if t.isalpha()]

    # Remove all stop words: no_stops
    no_stops = [t for t in alpha_only if t not in english_stops]
    articles.append(no_stops)



```

```python
# Importamos Diccionario
from gensim.corpora.dictionary import Dictionary

# Creamos un diccionario de gensim con los tokens en articles
dictionary = Dictionary(articles)

# Seleccionamos el id para "computer": computer_id
computer_id = dictionary.token2id.get("computer")
print(computer_id)

# Utiliza computer_id con el diccionario para imprimir la palabra
print(dictionary.get(computer_id))

# Utilizamos una lista de comprehension para iterar sobre el 'articulos' para crear un MmCorpus de gensim
# MmCorpus es la abreviatura de Matrix Market Corpus. Matrix Market es un format de almacenamiento e intercambio de matrices basado en ascii
corpus = [dictionary.doc2bow(article) for article in articles]

# Imprime las 10 primeras 
print(corpus[1][:10])
```

### 3.4 NER - Named Entity Recognition

En esta parte del ejercicio vamos a realizar un poco de Named-Entity Recognition sobre un artículo que vamos a importar utilizando NER. Para ello vamos a realizar los siguientes puntos.




1.   Convertir el artículo en frases.
2.   Convertir cada frase en palabras utilizando una comprehension list.
3.   Dentro de la comprehension list, etiquetar cada frase tokenizada en "Part of speech tagging" utilizando `nltk.pos_tag()`.
4.   Dividir cada frase etiquetada en named-entity chunks utilizando `nltk.ne_chunk_sents()`. Junto con `pos_sentences`, especifica el argumento de palabra clave adicional `binary=True`.



Nota: Antes de utilizar NER a través de NLTK, debes instalar los paquetes `averaged_perceptron_tagger`, `maxent_ne_chunker`, `words`

```python
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
#Abrimos el artículo
with open('/content/drive/MyDrive/Edem/NLP/News articles/uber_apple.txt', 'r') as file:
    article = file.read()
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize

sentences = sent_tokenize(article)

token_sentences = [word_tokenize(sent) for sent in sentences]

pos_sentences = [nltk.pos_tag(sent) for sent in token_sentences]

chunked_sentences = nltk.ne_chunk_sents(pos_sentences, binary=True)

for sent in chunked_sentences:
    for chunk in sent:
        if hasattr(chunk, "label") and chunk.label() == 'NE':
            print(chunk)
```


En este ejercicio, utilizarás algunas Named Entities extraídas y sus agrupaciones de una serie de artículos de periódicod para mostrar los distintos tipos de NER en los artículos.

Utilizarás un `defaultdict` llamado `ner_categories`, con claves que representan cada tipo de grupo de NE, y valores para contar el número de cada tipo de entidad con nombre diferente. Utilizarás una lista de frases fragmentadas llamada `chunked_sentences` similar a la del último ejercicio, pero esta vez con nombres de **categorías no binarias**.

Puedes utilizar `hasattr()` para determinar si cada chunk tiene una 'etiqueta' y luego simplemente utilizar el método `.label()` del chunk como la clave del diccionario.




```python
#Creamos la lista de frases fragmentadas
chunked_sentences = nltk.ne_chunk_sents(pos_sentences, binary=False)
print(chunked_sentences)
```

```python
#Creamos el defaultdict
ner_categories = defaultdict(int)

# Inicializamos el bucle anidado
for sent in chunked_sentences:
    for chunk in sent:
        if hasattr(chunk, 'label'):
            ner_categories[chunk.label()] += 1
            
# Crear una lista a partir del diccionario de claves para las labels del gráfico
labels = list(ner_categories.keys())

# Creamos una lista de valores
values = [ner_categories.get(v) for v in labels]

# Creamos el gráfico de tarta
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140)
plt.show()
```

## 4 Construyendo un Bot de Q&A Con BERT

En esta última parte de las prácticas, vamos a construir un sistema para Preguntas y respuestas basado en el modelo BERT, que viene de *Bidirectional Encoder Representations from Transformers*. BERT es una técnica de representación de texto como Word Embeddings.

<br>
BERT utiliza **Transformers**, un mecanismo de atención que aprende las relaciones contextuales entre las palabras (o subpalabras) de un texto. En su forma simple, Transformer incluye dos mecanismos separados: un Encoder (codificador) que lee el texto de entrada y un Decoder (decodificador) que produce una predicción para la tarea. Dado que el objetivo de BERT es generar un modelo lingüístico, **sólo es necesario el mecanismo codificador**. El funcionamiento detallado de Transformer se describe en [este paper de google](https://arxiv.org/pdf/1706.03762.pdf)

<br><br>

A diferencia de los modelos direccionales, que leen el input del texto secuencialmente (de izquierda a derecha o de derecha a izquierda), el codificador Transformer lee toda la secuencia de palabras a la vez. Por eso se considera bidireccional, aunque sería más exacto decir que es No direccional. Esta característica permite al modelo aprender el contexto de una palabra basándose en todo su entorno (a izquierda y a derecha de la palabra).

<br>



*BERT* fue desarrollado por Google en 2018 y ha demostrado funcionar muy bien para una variedad de tareas de NLP como la clasificación de textos, el resumen de textos, la generación de textos, machine translation, Q&A etc.
