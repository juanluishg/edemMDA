```python
from simpletransformers.seq2seq import Seq2SeqModel


model = Seq2SeqModel(
    "bert",
    "bert-base-multilingual-cased", #"outputs/encoder",
    "bert-base-multilingual-cased"  #"outputs/decoder" 
)
```

```python
#train_data - Pandas DataFrame containing the 2 columns - input_text, target_text
import pandas as pd

#input = ['¿dónde esta?','¿con quién está?','¿a qué hora?']
#output = ['En la cocina','Con Juan','Pronto']
#df_train = pd.DataFrame(data={'input_text': input, 'target_text': output}) ##to include eval_data=x

df_train = pd.read_csv("https://krono.act.uji.es/NLP-course/chitchat-friendly-es.csv")

df_train.rename(columns={'Question':'input_text','Answer':'target_text'}, inplace=True)

df_train.head()
```

```python
model.train_model(df_train, args = {"num_train_epochs": 1, "overwrite_output_dir": True, "save_eval_checkpoint": False, "save_model_every_epoch": False, "save_steps": -1})
```

```python
to_predict = ['¿cómo te llamas?','¿cuándo cumples años?','¿eres muy viejo?','¿Conoces a Alexa?','¡Se te ve muy bien!']

predictions = model.predict(to_predict)
print(predictions)
```
