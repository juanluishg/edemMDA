## AutoML

```python
!pip install h2o
```

```python
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

## Load Dataset

```python
titanic = pd.read_csv("./datasets/titanic.csv")
```

```python
train, test = train_test_split(titanic, train_size=0.75, random_state=42)
```

### h2o

```python
train_h2o = h2o.H2OFrame(train)
test_h2o = h2o.H2OFrame(test)
```

```python
x = train_h2o.columns
y = "Survived"
x.remove(y)
```

```python
train_h2o[y] = train_h2o[y].asfactor()
test_h2o[y] = test_h2o[y].asfactor()
```

```python
aml = H2OAutoML(max_models=20, seed=1)
aml.train(x=x, y=y, training_frame=train_h2o)
```

```python
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
aml.leader
```

```python
preds = aml.predict(test_h2o)
```

```python
metalearner = h2o.get_model("StackedEnsemble_BestOfFamily_AutoML_20210314_162756")
```

```python
perf = metalearner.model_performance(test_h2o)
perf.plot(type = "roc")
```

```python
metalearner.confusion_matrix(test_h2o)
```

```python
print("Model {} has an AUC of {}".format(metalearner.full_parameters["model_id"]["actual_value"]["name"], metalearner.auc(test_h2o)))
```

### Auto-SKLearn

```python
!pip install auto-sklearn
```

```python
dummies = pd.get_dummies(train["Sex"])
X_train = pd.concat([train, dummies], axis=1)
dummies = pd.get_dummies(test["Sex"])
X_test = pd.concat([test, dummies], axis=1)
```

```python
dummies = pd.get_dummies(train["Embarked"])
X_train = pd.concat([train, dummies], axis=1)
dummies = pd.get_dummies(test["Embarked"])
X_test = pd.concat([test, dummies], axis=1)
```

```python
X_train = train.drop(["Survived", "Name", "Ticket","Sex", "Cabin", "Embarked"], axis=1)
y_train = train["Survived"]

X_test = test.drop(["Survived", "Name", "Ticket","Sex", "Cabin", "Embarked"], axis=1)
y_test = test["Survived"]
```

```python
import autosklearn.classification

import sklearn.model_selection

import sklearn.datasets

import sklearn.metrics

automl = autosklearn.classification.AutoSklearnClassifier()

automl.fit(X_train, y_train)

y_hat = automl.predict(X_test)

print("Accuracy score", sklearn.metrics.accuracy_score(y_test, y_hat))
```

```python
print("Model {} has an AUC of {}".format(str(automl.get_models_with_weights()[0][1]).split("(")[0], metrics.roc_auc_score(y_test, y_hat)))
```

### TPot

```python
from tpot import TPOTClassifier
```

```python
dummies = pd.get_dummies(train["Sex"])
X_train = pd.concat([train, dummies], axis=1)
dummies = pd.get_dummies(test["Sex"])
X_test = pd.concat([test, dummies], axis=1)
```

```python
dummies = pd.get_dummies(train["Embarked"])
X_train = pd.concat([train, dummies], axis=1)
dummies = pd.get_dummies(test["Embarked"])
X_test = pd.concat([test, dummies], axis=1)
```

```python
X_train = train.drop(["Survived", "Name", "Ticket","Sex", "Cabin", "Embarked"], axis=1)
y_train = train["Survived"]

X_test = test.drop(["Survived", "Name", "Ticket","Sex", "Cabin", "Embarked"], axis=1)
y_test = test["Survived"]

tpot = TPOTClassifier(generations=50, population_size=50, verbosity=2, random_state=42, n_jobs=-1)
tpot.fit(X_train, y_train)
print(tpot.score(X_test, y_test))
tpot.export('tpot_digits_pipeline.py')
```

```python
y_pred = tpot.predict_proba(X_test)[:,1]
```

```python
print("Model {} has an AUC of {}".format(tpot.fitted_pipeline_, metrics.roc_auc_score(y_test, y_pred)))
```
